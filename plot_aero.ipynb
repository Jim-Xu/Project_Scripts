```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
f = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/sgm72/urban_plume_process.nc'
data =nc.Dataset(f)
```

```python
time = data.variables['time'][:]
```

```python
num_conc = data.variables['tot_num_conc'][:]
tot_bc_mass = data.variables['tot_bc_mass_conc'][:]
tot_oc_mass = data.variables['tot_oc_mass_conc'][:]
tot_mass = data.variables['tot_mass_conc'][:]
tot_dry_mass = data.variables['tot_dry_mass_conc'][:]
tot_so4_mass = data.variables['tot_so4_mass_conc'][:]
tot_no3_mass = data.variables['tot_no3_mass_conc'][:]
tot_nh4_mass = data.variables['tot_nh4_mass_conc'][:]
```

```python
plt.plot(time/3600,num_conc, color='green')
plt.ylim(5e9, 3e10)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Number Concentration ($\mathrm{m}^3$)")
plt.legend(loc='upper right')
plt.show()
```

```python
plt.plot(time/3600,tot_mass*1e9, color='green')
plt.plot(time/3600,tot_dry_mass*1e9, color='green',ls='--')

plt.ylim(0, 210)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Mass Concentration ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend(['Wet Mass','Dry Mass'])
plt.show()
```

```python
plt.plot(time/3600,tot_mass*1e9 - tot_dry_mass*1e9, color='green')
plt.ylim(0, 210)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("H$_2$O Concentration ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend()
plt.show()
```

```python
plt.plot(time/3600,tot_bc_mass*1e9, color='green')
plt.ylim(0, 10)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("BC Mass ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend()
plt.show()
```

```python
plt.plot(time/3600,tot_oc_mass*1e9, color='green')
plt.ylim(0, 20)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Total OC Mass ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend()
plt.show()
```

```python
plt.plot(time/3600,tot_so4_mass*1e9, color='green')
plt.ylim(0, 10)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Total $\mathrm{SO_4}$ Mass ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend()
plt.show()
```

```python
plt.plot(time/3600,tot_no3_mass*1e9, color='green')
plt.ylim(0, 30)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Total $\mathrm{NO_3}$ Mass ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend()
plt.show()
```
