```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
f_original = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_original/urban_plume_process.nc'
f_more_aerosol = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_more_aerosol_only/urban_plume_process.nc'
f_more_bc = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_more_bc/urban_plume_process.nc'
f_more_no2 = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_more_no2_only/urban_plume_process.nc'
f_two_change = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_more_aerosol_and_no2/urban_plume_process.nc'
data_original =nc.Dataset(f_original)
data_more_aerosol = nc.Dataset(f_more_aerosol)
data_more_bc =nc.Dataset(f_more_bc)
data_more_no2 =nc.Dataset(f_more_no2)
data_two_change =nc.Dataset(f_two_change)
```

```python
time = data_original.variables['time'][:]
```

```python
num_conc_original = data_original.variables['tot_num_conc'][:]
tot_bc_mass_original = data_original.variables['tot_bc_mass_conc'][:]
tot_oc_mass_original = data_original.variables['tot_oc_mass_conc'][:]
tot_mass_original = data_original.variables['tot_mass_conc'][:]
tot_dry_mass_original = data_original.variables['tot_dry_mass_conc'][:]
tot_so4_mass_original = data_original.variables['tot_so4_mass_conc'][:]
tot_no3_mass_original = data_original.variables['tot_no3_mass_conc'][:]
tot_nh4_mass_original = data_original.variables['tot_nh4_mass_conc'][:]
```

```python
num_conc_more_aerosol = data_more_aerosol.variables['tot_num_conc'][:]
tot_bc_mass_more_aerosol = data_more_aerosol.variables['tot_bc_mass_conc'][:]
tot_oc_mass_more_aerosol = data_more_aerosol.variables['tot_oc_mass_conc'][:]
tot_mass_more_aerosol = data_more_aerosol.variables['tot_mass_conc'][:]
tot_dry_mass_more_aerosol = data_more_aerosol.variables['tot_dry_mass_conc'][:]
tot_so4_mass_more_aerosol = data_more_aerosol.variables['tot_so4_mass_conc'][:]
tot_no3_mass_more_aerosol = data_more_aerosol.variables['tot_no3_mass_conc'][:]
tot_nh4_mass_more_aerosol = data_more_aerosol.variables['tot_nh4_mass_conc'][:]
```

```python
num_conc_more_bc = data_more_bc.variables['tot_num_conc'][:]
tot_bc_mass_more_bc = data_more_bc.variables['tot_bc_mass_conc'][:]
tot_oc_mass_more_bc = data_more_bc.variables['tot_oc_mass_conc'][:]
tot_mass_more_bc = data_more_bc.variables['tot_mass_conc'][:]
tot_dry_mass_more_bc = data_more_bc.variables['tot_dry_mass_conc'][:]
tot_so4_mass_more_bc = data_more_bc.variables['tot_so4_mass_conc'][:]
tot_no3_mass_more_bc = data_more_bc.variables['tot_no3_mass_conc'][:]
tot_nh4_mass_more_bc = data_more_bc.variables['tot_nh4_mass_conc'][:]
```

```python
num_conc_more_no2 = data_more_no2.variables['tot_num_conc'][:]
tot_bc_mass_more_no2 = data_more_no2.variables['tot_bc_mass_conc'][:]
tot_oc_mass_more_no2 = data_more_no2.variables['tot_oc_mass_conc'][:]
tot_mass_more_no2 = data_more_no2.variables['tot_mass_conc'][:]
tot_dry_mass_more_no2 = data_more_no2.variables['tot_dry_mass_conc'][:]
tot_so4_mass_more_no2 = data_more_no2.variables['tot_so4_mass_conc'][:]
tot_no3_mass_more_no2 = data_more_no2.variables['tot_no3_mass_conc'][:]
tot_nh4_mass_more_no2 = data_more_no2.variables['tot_nh4_mass_conc'][:]
```

```python
num_conc_two_change = data_two_change.variables['tot_num_conc'][:]
tot_bc_mass_two_change = data_two_change.variables['tot_bc_mass_conc'][:]
tot_oc_mass_two_change = data_two_change.variables['tot_oc_mass_conc'][:]
tot_mass_two_change = data_two_change.variables['tot_mass_conc'][:]
tot_dry_mass_two_change = data_two_change.variables['tot_dry_mass_conc'][:]
tot_so4_mass_two_change = data_two_change.variables['tot_so4_mass_conc'][:]
tot_no3_mass_two_change = data_two_change.variables['tot_no3_mass_conc'][:]
tot_nh4_mass_two_change = data_two_change.variables['tot_nh4_mass_conc'][:]
```

```python
plt.plot(time/3600,num_conc_original, color='green', label='test_original')
plt.plot(time/3600,num_conc_more_bc, color='red', label='test_more_bc')
plt.plot(time/3600,num_conc_more_aerosol, color='black', label='test_more_aerosol')
plt.plot(time/3600,num_conc_more_no2, color='orange', label='test_more_no2')
plt.plot(time/3600,num_conc_two_change, color='purple', label='test_two_change')
plt.ylim(5e9, 3e10)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Number Concentration ($\mathrm{m}^3$)")
plt.legend(['Original', 'More BC', 'More Aerosol' ,'More NO$_2$', 'Two Changes'], loc='upper right')
plt.show()
```

```python
plt.plot(time/3600,tot_mass_original*1e9, color='green', label='test_original')
plt.plot(time/3600,tot_dry_mass_original*1e9, color='green', label='test_original',ls='--')
plt.plot(time/3600,tot_mass_more_bc*1e9, color='red', label='test_more_bc')
plt.plot(time/3600,tot_dry_mass_more_bc*1e9, color='red', label='test_more_bc',ls='--')
plt.plot(time/3600,tot_mass_more_aerosol*1e9, color='black', label='test_more_aerosol')
plt.plot(time/3600,tot_dry_mass_more_aerosol*1e9, color='black', label='test_more_aerosol',ls='--')
plt.plot(time/3600,tot_mass_more_no2*1e9, color='orange', label='test_more_no2')
plt.plot(time/3600,tot_dry_mass_more_no2*1e9, color='orange', label='test_more_no2',ls='--')
plt.plot(time/3600,tot_mass_two_change*1e9, color='purple', label='test_two_change')
plt.plot(time/3600,tot_dry_mass_two_change*1e9, color='purple', label='test_two_change',ls='--')
plt.ylim(0, 210)
plt.xlim(0, 25)
plt.grid()
plt.xlabel("Time (h)")
plt.ylabel("Mass Concentration ($\mu \mathrm{g}/\mathrm{m}^3$)")
plt.legend(['Wet Mass','Dry Mass'])
# plt.legend(['Original', 'More BC', 'More Aerosol' ,'More NO$_2$', 'Two Changes'], loc='upper left')
plt.show()
```
