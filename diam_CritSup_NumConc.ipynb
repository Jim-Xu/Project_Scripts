```python
import numpy as np
import netCDF4 as nc
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
file = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_original/urban_plume_00000015_process.nc'
ncf =nc.Dataset(file)
```

```python
diam_edges = ncf.variables["diam_edges"][:] * 1e6
sc_edges = ncf.variables["sc_edges"][:] * 100
diam_sc_dist = ncf.variables["diam_sc_dist"][:] * 1e-6
```

```python
sc = plt.pcolormesh(diam_edges, sc_edges, diam_sc_dist,
                    norm = colors.LogNorm(vmin=1e3, vmax=1e5), linewidths = 0.1)
plt.xscale("log")
plt.xlabel(r"Dry Diameter $D_{\rm dry}$ / $\rm \mu m$")
plt.xlim(1e-2, 1e0)

plt.yscale("log")
plt.ylabel(r"critical supersaturation $S_{\rm c}$ / \%")
plt.ylim(1e-2, 1e2)

plt.grid(True)
plt.colorbar(sc, label=r"Number Conc. $n(D_{\rm dry},w_{\rm BC})$ / $\rm cm^{-3}$")
```
