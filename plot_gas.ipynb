```python
import sys, os
import matplotlib
import scipy.io, numpy
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import netCDF4
import matplotlib as mpl
import seaborn as sns
```

```python
time = np.zeros(25)
```

```python
gas_mixing_ratio_ori = np.zeros((25, 77))
directory_ori = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_original'
for subdir, dirs, files in os.walk(directory_ori):
    for file in files:
        filename = subdir + os.sep + file
        if ("urban_plume_00000" in filename) and (".nc") in filename:
            print(filename[-5:-3])
            i = int(filename[-5:-3])
            ncf = scipy.io.netcdf_file(filename, mmap=False)
            gas_mixing_ratio_ori[i-1,:] = ncf.variables["gas_mixing_ratio"].data
            time[i-1] =  ncf.variables["time"].data/3600
```

```python
gas_mixing_ratio_bc = np.zeros((25, 77))
directory_bc = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_more_bc'
for subdir, dirs, files in os.walk(directory_bc):
    for file in files:
        filename = subdir + os.sep + file
        if ("urban_plume_0001" in filename) and (".nc") in filename:
            i = int(filename[-5:-3])
            ncf = scipy.io.netcdf_file(filename, mmap=False)
            gas_mixing_ratio_bc[i-1,:] = ncf.variables["gas_mixing_ratio"].data
            time[i-1] =  ncf.variables["time"].data/3600
```

```python
plt.plot(gas_mixing_ratio_ori[:,10], color='red')
plt.plot(gas_mixing_ratio_bc[:,10], color='purple')
plt.plot(gas_mixing_ratio_aero[:,10], color='green')
plt.plot(gas_mixing_ratio_no2[:,10], color='black')
plt.plot(gas_mixing_ratio_aero_no2[:,10], color='orange')
plt.grid()
plt.xlim(0,25)
plt.ylim(0,130)
plt.xlabel("Time (h)")
plt.ylabel('$O_3$ mixing ratio (ppb)')
plt.legend(['original','more_bc','more_aero','more_no2','two change'])
```
