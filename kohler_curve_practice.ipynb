```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 288
R = 8.314 
sigma = 7.2* 1e-2 
mv = 18 * 1e-3
rho_l = 1000 
T_0 = 273.15 
r = np.logspace(-8,-5,100) 
```

```python
def es_flat(r, T):
    T_c= T - T_0
    es_T = 611.2 * np.exp(17.67*T_c/(T_c+243.5))
    return es_T*np.exp(2*sigma*mv/(R*rho_l*r*T))
```

```python
#Q1
#Saturation vapor pressure over a curved surface
plt.xscale('log')
plt.plot(r*1e6, es_flat(r, T))
plt.grid()
plt.xlabel("Diameter ($\mu$m)")
plt.ylabel("saturation vapor pressure (Pa)")
```

```python
#Q2
#When the water droplet has 50 nm
plt.xscale('log')
plt.plot(r*1e6, es_flat(r, T))
plt.axvline(5e-2, color='red')
plt.axhline(1750, color='red')
plt.grid()
plt.xlabel("Diameter ($\mu$m)")
plt.ylabel("saturation vapor pressure (Pa)")
```

```python
#when radius is 50nm and pressure is 1750Pa, supersaturation vapor pressure is greater, so this droplet will growth.
```

```python
#Q3
def sat_vap_flat(T):
    T_c= T - T_0
    es_T = 611.2*np.exp(17.67*T_c/(T_c+243.5))
    return es_T
sat_vap_flat(273.15)
```

```python
#Q4
s = es_flat(r, T)/sat_vap_flat(T)

for T in [200, 250, 270, 300]:
    plt.xscale('log')
    plt.plot(r*1e6,es_flat(r, T)/sat_vap_flat(T), label='T=%d K'%T)
    plt.legend(loc=1)
plt.grid()
plt.xlabel("Diameter ($\mu$m)")
plt.ylabel("Saturation Pressure Ratio")
```

```python
#Q5
#(1) for NcCl
i = 2
M = 58.5 * 1e-3
T = 288 
R = 8.314 
sigma = 7.2*(10**-2) 
mv = 18 * 1e-3
rho_l = 1000 
r = np.logspace(-8,-4,100) 
```

```python
 def S(i,m,M,r):
    b = 2*sigma*mv/(rho_l*R*T)
    c = 3*i*m*1e-3*mv/(4*np.pi*rho_l*M)
    return (1-c/(r**3))*np.exp(b/r)
```

```python
i = 2
M = 58.5 * 1e-3
for m in [1e-12, 1e-13, 1e-14, 1e-15, 1e-16]:
    plt.xscale('log')
    plt.plot(r*1e6, (S(i,m,M,r)-1)*100, label='m=%.e g'%m)
    plt.legend(loc=4)

plt.ylim(-1.5, 0.5)
plt.xlim(0.01, 100)
plt.grid()
plt.xlabel("Diameter ($\mu$m)")
plt.ylabel("Superaturation (%)")
```

```python
#(2) for (NH4)2SO4
i = 3
M = 132.14 * 1e-3

for m in [1e-12, 1e-13, 1e-14, 1e-15, 1e-16]:
    plt.xscale('log')
    plt.plot(r*1e6, (S(i,m,M,r)-1)*100, label='m=%.e g'%m)
    plt.legend(loc=4)
    
plt.ylim(-1.5, 0.5)
plt.xlim(0.01, 100)
plt.grid()
plt.xlabel("Diameter ($\mu$m)")
plt.ylabel("Superaturation (%)")
```
