```python
import numpy as np
import netCDF4 as nc
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
file = '/data/keeling/a/xx24/d/partmc/scenarios/1_urban_plume/out_original/urban_plume_00000015_process.nc'
ncf =nc.Dataset(file)
```

```python
diam_edges = ncf.variables["diam_edges"][:]* 1e6
bc_frac_edges = ncf.variables["bc_frac_edges"][:] * 100
diam_bc_dist = ncf.variables["diam_bc_dist"][:] * 1e-6
```

```python
bc = plt.pcolormesh(diam_edges, bc_frac_edges, diam_bc_dist,
                    norm = colors.LogNorm(vmin=1e3, vmax=1e5), linewidths = 0.1)
plt.xscale("log")
plt.xlabel(r"Dry Diameter $D_{\rm dry}$ / $\rm \mu m$")
plt.xlim(1e-2, 1e0)

plt.yscale("linear")
plt.ylabel(r"BC Mass Fraction $w_{\rm BC}$ / %")
plt.ylim(0, 80)

plt.grid(True)
plt.colorbar(bc, label=r"Number Conc. $n(D_{\rm dry},w_{\rm BC})$ / $\rm cm^{-3}$")
```
